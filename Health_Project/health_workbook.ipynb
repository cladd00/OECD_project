```python
x1 = pd.ExcelFile('Health_Data.xlsx')
print(x1.sheet_names)

health = x1.parse('Health_Exp')
smoker = x1.parse('Smoker')
life_exp = x1.parse('Life Exp')
hospital = x1.parse('Hospital')
infant = x1.parse('Infant Mortality')
doctor = x1.parse('Doctor')
```

```python
df = smoker.set_index('LOCATION')
df = df.loc['USA']

df2 = smoker.set_index('LOCATION')
df2 = df2.loc['CHE']

df3 = smoker.set_index('LOCATION')
df3 = df3.loc['CAN']

df4=smoker.set_index('LOCATION')
df4 = df4.loc['DEU']

df_new = pd.concat([df,df2])


df_new_new = pd.concat([df_new,df3])


df_new_new = pd.concat([df_new_new,df4])
df_new_new.head()
```

```python
df = df_new_new.reset_index()
usa_tot = df[(df['LOCATION'] =='USA') & (df['SUBJECT'] == 'TOT')]
usa_tot_x = usa_tot.TIME
usa_tot_y = usa_tot.Value


can_tot = df[(df['LOCATION'] =='CAN') & (df['SUBJECT'] == 'TOT')]
can_tot_x = can_tot.TIME
can_tot_y = can_tot.Value


che_tot = df[(df['LOCATION'] =='CHE')& (df['SUBJECT'] == 'TOT')]
che_tot_x = che_tot.TIME
che_tot_y = che_tot.Value

deu_tot = df[(df['LOCATION'] =='DEU')& (df['SUBJECT'] == 'TOT')]
deu_tot_x = deu_tot.TIME
deu_tot_y = deu_tot.Value

usa_men = df[(df['LOCATION'] =='USA') & (df['SUBJECT'] == 'MEN')]
usa_men_x = usa_men.TIME
usa_men_y = usa_men.Value


can_men = df[(df['LOCATION'] =='CAN') & (df['SUBJECT'] == 'MEN')]
can_men_x = can_men.TIME
can_men_y = can_men.Value


che_men = df[(df['LOCATION'] =='CHE')& (df['SUBJECT'] == 'MEN')]
che_men_x = che_men.TIME
che_men_y = che_men.Value

deu_men = df[(df['LOCATION'] =='DEU')& (df['SUBJECT'] == 'MEN')]
deu_men_x = deu_men.TIME
deu_men_y = deu_men.Value

usa_women = df[(df['LOCATION'] =='USA') & (df['SUBJECT'] == 'WOMEN')]
usa_women_x = usa_women.TIME
usa_women_y = usa_women.Value


can_women = df[(df['LOCATION'] =='CAN') & (df['SUBJECT'] == 'WOMEN')]
can_women_x = can_women.TIME
can_women_y = can_women.Value


che_women = df[(df['LOCATION'] =='CHE')& (df['SUBJECT'] == 'WOMEN')]
che_women_x = che_women.TIME
che_women_y = che_women.Value

deu_women = df[(df['LOCATION'] =='DEU')& (df['SUBJECT'] == 'WOMEN')]
deu_women_x = deu_women.TIME
deu_women_y = deu_women.Value
```

```python
import numpy as np
import matplotlib.pyplot as plt

labels = ['United States','Canada','Switzerland','Germany']
plt.scatter(usa_tot_x,usa_tot_y,color ='red')
plt.scatter(can_tot_x,can_tot_y,color ='blue')
plt.scatter(che_tot_x,che_tot_y,color = 'green')
plt.scatter(deu_tot_x,deu_tot_y,color = 'purple')
plt.legend(labels = labels)
plt.title('Smoker TOT')
plt.xlabel('Years')
plt.ylabel('Percentage')

plt.show()
```

```python
df = health.set_index('LOCATION')
df = df.loc['USA']

df2 = health.set_index('LOCATION')
df2 = df2.loc['CHE']

df3 = health.set_index('LOCATION')
df3 = df3.loc['CAN']

df4=health.set_index('LOCATION')
df4 = df4.loc['DEU']

df_new = pd.concat([df,df2])


df_new_new = pd.concat([df_new,df3])


df_new_new = pd.concat([df_new_new,df4])
df_new_new.head()
```

```python
df_n = df_new_new.reset_index()

can_tot = df_n[(df_n['SUBJECT'] == 'TOT') & (df_n['LOCATION'] == 'CAN') & (df_n['MEASURE'] =='PC_GDP')]
can_comp = df_n[(df_n['SUBJECT'] == 'COMPULSORY') & (df_n['LOCATION'] == 'CAN')& (df_n['MEASURE'] =='PC_GDP')]
can_vol = df_n[(df_n['SUBJECT'] == 'VOLUNTARY') & (df_n['LOCATION'] == 'CAN')& (df_n['MEASURE'] =='PC_GDP')]


can_tot = can_tot.set_index('LOCATION')
can_comp = can_comp.set_index('LOCATION')
can_vol = can_vol.set_index('LOCATION')

usa_tot = df_n[(df_n['SUBJECT'] == 'TOT') & (df_n['LOCATION'] == 'USA')& (df_n['MEASURE'] =='PC_GDP')]
usa_comp = df_n[(df_n['SUBJECT'] == 'COMPULSORY') & (df_n['LOCATION'] == 'USA')& (df_n['MEASURE'] =='PC_GDP')]
usa_vol = df_n[(df_n['SUBJECT'] == 'VOLUNTARY') & (df_n['LOCATION'] == 'USA')& (df_n['MEASURE'] =='PC_GDP')]

usa_tot = usa_tot.set_index('LOCATION')
usa_comp = usa_comp.set_index('LOCATION')
usa_vol = usa_vol.set_index('LOCATION')


che_tot = df_n[(df_n['SUBJECT'] == 'TOT') & (df_n['LOCATION'] == 'CHE')& (df_n['MEASURE'] =='PC_GDP')]
che_comp = df_n[(df_n['SUBJECT'] == 'COMPULSORY') & (df_n['LOCATION'] == 'CHE')& (df_n['MEASURE'] =='PC_GDP')]
che_vol = df_n[(df_n['SUBJECT'] == 'VOLUNTARY') & (df_n['LOCATION'] == 'CHE')& (df_n['MEASURE'] =='PC_GDP')]

che_tot = che_tot.set_index('LOCATION')
che_comp =che_comp.set_index('LOCATION')
che_vol = che_vol.set_index('LOCATION')

deu_tot = df_n[(df_n['SUBJECT'] == 'TOT') & (df_n['LOCATION'] == 'DEU')& (df_n['MEASURE'] =='PC_GDP')]
deu_comp = df_n[(df_n['SUBJECT'] == 'COMPULSORY') & (df_n['LOCATION'] == 'DEU')& (df_n['MEASURE'] =='PC_GDP')]
deu_vol = df_n[(df_n['SUBJECT'] == 'VOLUNTARY') & (df_n['LOCATION'] == 'DEU')& (df_n['MEASURE'] =='PC_GDP')]

deu_tot = deu_tot.set_index('LOCATION')
deu_comp =deu_comp.set_index('LOCATION')
deu_vol = deu_vol.set_index('LOCATION')
```

```python
usa_tot_x = usa_tot.TIME
usa_tot_y = usa_tot.Value
usa_comp_x = usa_comp.TIME
usa_comp_y = usa_comp.Value
usa_vol_x = usa_vol.TIME
usa_vol_y = usa_vol.Value


can_tot_x = can_tot.TIME
can_tot_y = can_tot.Value
can_comp_x = can_comp.TIME
can_comp_y = can_comp.Value
can_vol_x = can_vol.TIME
can_vol_y = can_vol.Value


che_tot_x = che_tot.TIME
che_tot_y = che_tot.Value
che_comp_x = che_comp.TIME
che_comp_y = che_comp.Value
che_vol_x = che_vol.TIME
che_vol_y = che_vol.Value

deu_tot_x = deu_tot.TIME
deu_tot_y = deu_tot.Value
deu_comp_x = deu_comp.TIME
deu_comp_y = deu_comp.Value
deu_vol_x = deu_vol.TIME
deu_vol_y = deu_vol.Value
```

```python
import numpy as np
import matplotlib.pyplot as plt

labels = ['United States','Canada','Switzerland','Germany']
plt.scatter(usa_tot_x,usa_tot_y,color ='red')
plt.scatter(can_tot_x,can_tot_y,color ='blue')
plt.scatter(che_tot_x,che_tot_y,color = 'green')
plt.scatter(deu_tot_x,deu_tot_y,color = 'purple')
plt.legend(labels = labels)
plt.title('Total Health Expenditure(%GDP)')
plt.xlabel('Years')
plt.ylabel('Percentage')

plt.show()
```
